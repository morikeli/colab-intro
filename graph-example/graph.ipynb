** Plotting a simple graphs with two arrays **

```python
import matplotlib.pyplot as plt
import numpy as np

# We have to import the matplotlib module and numpy library as most of the data we work withwill be in the form of arrays only.
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('First graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# first array will appear on the x-axis and the second array will appear on the y-axis

plt.show()
```
