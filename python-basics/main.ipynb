** Define a changeable tuple **

```python
cn = ('yi', 'er', 'san')
en = ('one', 'two', 'three')
num = (1, 2, 3)

tmp = [cn, en, num, [1.1, 2.2], 'language']
print(tmp)
print(tmp[0][0])
print(tmp[0][0][0])
```
